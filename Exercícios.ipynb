Um programa deve ser escrito para ler dois números e, em seguida, perguntar à pessoa usuária qual operação ele deseja realizar. O resultado da operação deve incluir informações sobre o número - se é par ou ímpar, positivo ou negativo e inteiro ou decimal.

A partir do valor da variação, deve ser enviada às seguintes sugestões:
- Para variação acima de 20%: bonificação para o time de vendas.

#### Questão 8
Vamos entender a distribuição de idades de pensionistas de uma empresa de previdência. Tendo esses valores, faça um código que gere uma lista contendo o percentual de crescimento de bactérias por dia, comparando o número de bactérias em cada dia com o número de bactérias do dia anterior. Dica: para calcular o percentual de crescimento usamos a seguinte equação: `100 * (amostra_atual - amostra_passada) / (amostra_passada)`.

````python
lista_bacterias = [1.2, 2.1, 3.3, 5.0, 7.8, 11.3, 16.6, 25.1, 37.8, 56.9]

lista_porcentagem = []

for dia in range(1, len(lista_bacterias)):
  porcentagem = 100 * (lista_bacterias[dia] - lista_bacterias[dia - 1]) / lista_bacterias[dia - 1]
  lista_porcentagem.append(porcentagem)

print(lista_porcentagem)
````

````python
idades_geral = {'Setor A': [22, 26, 30, 30, 35, 38, 40, 56, 57, 65],
                'Setor B': [22, 24, 26, 33, 41, 49, 50, 54, 60, 64],
                'Setor C': [23, 26, 26, 29, 34, 35, 36, 41, 52, 56],
                'Setor D': [19, 20, 25, 27, 34, 39, 42, 44, 50, 65]}

soma_medias = 0
pessoas_acima_media = 0

for setor, idades in idades_geral.items():
  media_idade_setor = sum(idades) / 10
  print(f'A média de idade do {setor} é {media_idade_setor} anos')
  soma_medias += media_idade_setor

media_geral = soma_medias / 4
print(f'A média geral é de {media_geral} anos')

for idades in idades_geral.values():
  for id in idades:
    if id > media_geral:
      pessoas_acima_media += 1
print(f'{pessoas_acima_media} pessoas estão acima da média geral')
````
